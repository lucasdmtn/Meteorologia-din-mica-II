```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import ArtistAnimation
from IPython.display import Image
```

# Eq. Difusão

## $ \partial f/ \partial t = \alpha \partial ² f / \partial dx^2 $


```python
ci = [1,2,3,4,5,6,7,8,7,6,5,4,3,2,1]
nx = len(ci) #15
dx = 2
nt = 50
nu = 0.5*np.pi**2.0
dt = 0.2
plt.plot(ci)
plt.title('Condição inicial')
plt.xlabel('Ponto de grade')
plt.ylabel('Temperatura')
```

```python
fig, ax = plt.subplots()
ims=[]
for n in range(nt):  
    un = ci.copy()
    for i in range(1, nx - 1):
        # método explícito Avançado no Tempo Centrado no Espaço
        ci[i] = un[i] + nu * dt / dx**2 * (un[i+1] - 2 * un[i] + un[i-1])
    im = ax.plot(ci)
    ims.append(im)
    ax.set_title('Solução Eq. de Difusão')
    plt.xlabel('Ponto de grade')
    plt.ylabel('Temperatura')
    plt.close()
```

```python
g = ArtistAnimation(fig, ims)
g.save('solucao.gif')
Image(filename="./solucao.gif")
```
